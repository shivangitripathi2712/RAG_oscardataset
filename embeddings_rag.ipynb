```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
def text_embedding(text) -> None:    # Define a function that takes a piece of text and returns its vector embedding.
    model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')   # Instantiate a SentenceTransformer model using the 'all-MiniLM-L6-v2' pre-trained model.
    return model.encode(text,normalize_embeddings=True). # Return the normalized embedding of the input text.
```

```python
phrase1="Apple is a fruit"
embedding1=text_embedding(phrase1) # Define a phrase related to the concept of 'Apple' as a fruit.
```

```python
phrase2="Apple iPhone is expensive"
embedding2=text_embedding(phrase2)
```

```python
phrase3="Mango is a fruit"
embedding3=text_embedding(phrase3)
```

```python
phrase4="There is a new Apple iPhone"
embedding4=text_embedding(phrase4)
```

```python
def vector_similarity(vec1, vec2):
    return np.dot(np.squeeze(np.array(vec1)),np.squeeze(np.array(vec2)))
```

```python
print(vector_similarity(embedding1,embedding3))
print(vector_similarity(embedding1,embedding4)) # calculate the similarity between the embedding
```

```python
print(vector_similarity(embedding2,embedding3))
print(vector_similarity(embedding2,embedding4))
```
