```python
import pandas as pd
import chromadb
from sentence_transformers import SentenceTransformer
import numpy as np
import os.    #
```

```python
df=pd.read_csv('/content/oscars.csv')
df
```

```python
df=df.loc[df['year_ceremony'] == 2023]
df
```

```python
df=df.dropna(subset=['film'])
df.loc[:, 'category'] = df['category'].str.lower()
df # Drop any rows from the DataFrame where the 'film' column has NA/null data.
```

```python
df.loc[:, 'text'] = df['name'] + ' got nominated under the category, ' + df['category'] + ', for the film ' + df['film'] + ' to win the award'
df.loc[df['winner'] == False, 'text'] = df['name'] + ' got nominated under the category, ' + df['category'] + ', for the film ' + df['film'] + ' but did not win'
df # Create a new column 'text' in the DataFrame that combines the nominee's name, category, and film into a sentence.
```

```python
client = chromadb.Client() # Instantiate a Client object from the chromadb library.
collection = client.get_or_create_collection("oscars-2023") # Get an existing collection named "oscars-2023"
```

```python
docs=df["text"].tolist() # Convert the 'text' column of the DataFrame to a list and store it in 'docs'.
```

```python
ids= [str(x) for x in df.index.tolist()]# Convert the DataFrame index to a list of strings for unique document identifiers and store it in 'ids'
```

```python
collection.add(
    documents=docs,
    ids=ids
)
```

```python
results = collection.query(
    query_texts=["RRR"],
    n_results=1
) # Query the collection for documents related to the query text "RRR" and retrieve 1 result.
```
